```python
import pandas as pd

# Define the file path
file_path = r"C:\Users\u1246538\Downloads\liss-data\Notebooks_MHI-5\merged_dataset2022.csv"

# Load the CSV file into a DataFrame
merged_dataset2022 = pd.read_csv(file_path)

# Check the first few rows to ensure it's loaded correctly
print(merged_dataset2022.head())
```

```python
# Definieer de blokken (maak een lijst met de kolommen die je wilt voor elk blok)
demographic_block = ['age', 'gender', 'sex',  'education', 'Employment_status', 'Personal_Net_Income_Category', 'Ethnic_Background' , 'living_arrangement', 'marital_status']
health_block = ['SmallSnacks_Daily', 'LargeSnacks_Weekly', 'used_hallucinogens', 'Self-Rated_Health', 'Health_Hindrance_to_Daily_Functioning', 'Sleeping_Problems']
social_factors_block = ['hours_on_social_media', 'loneliness_score', 'social_media_frequency']
psychology_block = ['emotional_stability', 'intellect_imagination', 'conscientiousness' , 'agreeableness' , 'extraversion' , 'self_esteem', 'composite_values']

# Groeperen van de features in de vier blokken uit de 'merged_dataset2022'
df_demographic = merged_dataset2022[demographic_block]
df_health = merged_dataset2022[health_block]
df_social_factors = merged_dataset2022[social_factors_block]
df_psychology = merged_dataset2022[psychology_block]

# Combine different blocks
df_social_demographic = merged_dataset2022[demographic_block + social_factors_block]
df_socio_demo_health = merged_dataset2022[demographic_block + social_factors_block + health_block]
df_socio_demo_psychology = merged_dataset2022[demographic_block + social_factors_block + psychology_block]
df_socio_demo_health_psychology = merged_dataset2022[demographic_block + social_factors_block + health_block + psychology_block]

# Print the first few rows for verification
print("Demographic Block:\n", df_demographic.head())
print("\nHealth Block:\n", df_health.head())
print("\nSocial Factors Block:\n", df_social_factors.head())
print("\nPsychology Block:\n", df_psychology.head())

print("\nSocial + Demographic Block:\n", df_social_demographic.head())
print("\nSocio + Demo + Health Block:\n", df_socio_demo_health.head())
print("\nSocial + Demo + Psychology Block:\n", df_socio_demo_psychology.head())
print("\nSocio + Demo + Health + Psychology Block:\n", df_socio_demo_health_psychology.head())

# Optional: Save the combined blocks for future use
# df_social_demographic.to_csv("path_to_save/social_demographic_block.csv", index=False)
# df_socio_demo_health.to_csv("path_to_save/socio_demo_health_block.csv", index=False)
# df_socio_demo_psychology.to_csv("path_to_save/socio_demo_psychology_block.csv", index=False)
# df_socio_demo_health_psychology.to_csv("path_to_save/socio_demo_health_psychology_block.csv", index=False)


```

```python
# Define the target variable
target_column = ['mhi5_class_2022']  # Replace with actual target name

# Columns that are inside blocks but MAY only be needed for error analysis
possible_error_analysis_columns = ['gender', 'Religion', 'Religious_Membership', 'political_interest' , 'mhi5_std_score_2022']


# Full list of extra columns
keep_columns = target_column + possible_error_analysis_columns
```

```python
# Combine Social + Demographic blocks + target + extra columns
df_social_demographic = merged_dataset2022[demographic_block + social_factors_block + keep_columns]

# Combine Socio + Demo + Health blocks + target + extra columns
df_socio_demo_health = merged_dataset2022[demographic_block + social_factors_block + health_block + keep_columns]

# Combine Social + Demo + Psychology blocks + target + extra columns
df_socio_demo_psychology = merged_dataset2022[demographic_block + social_factors_block + psychology_block + keep_columns]

# Combine Socio + Demo + Health + Psychology blocks + target + extra columns
df_socio_demo_health_psychology = merged_dataset2022[demographic_block + social_factors_block + health_block + psychology_block + keep_columns]
```

```python
df_social_demographic.to_csv("social_demographic.csv", index=False)
df_socio_demo_health.to_csv("socio_demo_health.csv", index=False)
df_socio_demo_psychology.to_csv("socio_demo_psychology.csv", index=False)
df_socio_demo_health_psychology.to_csv("socio_demo_health_psychology.csv", index=False)
```
